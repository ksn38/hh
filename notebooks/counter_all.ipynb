```python
import re
from os import listdir
from collections import Counter
```

```python
mypath = '../luxoft/source/'
files = listdir(mypath)
words = []

for i in files:
    f = open(mypath + i, 'r', encoding='utf-8')
    t = f.read()
    t = t.replace('\n', ' ')
    t = t.split(' ')
    for i in t:
        i = re.sub('\W', ' ', i)
        i = i.split(' ')
        for j in i:
            words.append(j)
    f.close()

sorted(Counter(words).items(), key=lambda x: x[1], reverse=True)
        
```
